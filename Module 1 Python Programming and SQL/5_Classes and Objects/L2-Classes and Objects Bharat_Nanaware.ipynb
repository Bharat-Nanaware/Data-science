#### Create the classes and functions as per following description: 
- Create a class Person : 
   - The attributes of the class are name, age, city
   - The functions of the class are :
       - __init__(self, name, age, city)
       - show_details(self)
       - update_city(self, new_city)
       
- Create a class Employee which inherits the class Person:
    - The attributes of the class are name , age, city, salary, position, dept
    - The functions of the class are : 
        - __init__(self, name , age, city, salary, position, dept)
        - show_details(self)
        - update_salary(self, new_salary)
        - update_dept(self, new_dept)
        - update_position(self, new_pos)
        - inc_salary(self)
        - dec_salary
        
- Make use of the super function when necessary
- Create objects and Demonstrate working of all the functions

```python
class Person:
    def __init__(self, name, age, city):
        self.name = name
        self.age = age
        self.city = city
    
    def show_details(self):
        print("Name:", self.name)
        print("Age:", self.age)
        print("City:", self.city)
    
    def update_city(self, new_city):
        self.city = new_city


class Employee(Person):
    def __init__(self, name, age, city, salary, position, dept):
        super().__init__(name, age, city)
        self.salary = salary
        self.position = position
        self.dept = dept
    
    def show_details(self):
        super().show_details()
        print("Salary:", self.salary)
        print("Position:", self.position)
        print("Department:", self.dept)
    
    def update_salary(self, new_salary):
        self.salary = new_salary
    
    def update_dept(self, new_dept):
        self.dept = new_dept
    
    def update_position(self, new_pos):
        self.position = new_pos
    
    def inc_salary(self, amount):
        self.salary += amount
    
    def dec_salary(self, amount):
        self.salary -= amount


# Create objects and demonstrate the functions
person1 = Person("Alice", 30, "New York")
print("Details of Person:")
person1.show_details()
print()

person1.update_city("Los Angeles")
print("Updated City:", person1.city)
print()

employee1 = Employee("Bob", 35, "Chicago", 50000, "Manager", "HR")
print("Details of Employee:")
employee1.show_details()
print()

employee1.update_salary(55000)
print("Updated Salary:", employee1.salary)
print()

employee1.update_dept("Finance")
print("Updated Department:", employee1.dept)
print()

employee1.inc_salary(5000)
print("Increased Salary:", employee1.salary)
print()

employee1.dec_salary(3000)
print("Decreased Salary:", employee1.salary)
```
